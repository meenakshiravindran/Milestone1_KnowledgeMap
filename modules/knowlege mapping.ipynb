```python
import pandas as pd
import networkx as nx
 
# Load triples (from Day 5)
triples = pd.read_csv("triples.csv")
 
# Create a directed graph
G = nx.DiGraph()
 
# Add nodes and edges
for _, row in triples.iterrows():
    subj = row["Subject"]
    obj = row["Object"]
    rel = row["Relation"]
    G.add_node(subj)
    G.add_node(obj)
    G.add_edge(subj, obj, label=rel)
 
print(f"‚úÖ Graph built with {len(G.nodes())} nodes and {len(G.edges())} edges.")
 
# Example: print connections for one node
print("\nRelations for 'it':")
for neighbor in G["it"]:
    print("‚Üí", neighbor, "| Relation:", G["it"][neighbor]["label"])
```

```python
from pyvis.network import Network
 
# Create PyVis network
net = Network(height="600px", width="100%", directed=True, bgcolor="#222222", font_color="white")

G = nx.relabel_nodes(G, lambda x: str(x))
# Load NetworkX graph into PyVis
net.from_nx(G)
 
# Customize edges
for e in G.edges(data=True):
    net.add_edge(e[0], e[1], title=e[2]["label"], label=e[2]["label"])
 
# Save and open in browser
net.write_html("knowledge_graph.html", open_browser=False)
print("üåê Graph saved as knowledge_graph.html ‚Äî open it in your browser!")
```

```python
#main task
import pandas as pd
import networkx as nx
from pyvis.network import Network

data = pd.read_csv("tripless.csv")  

G = nx.DiGraph()

for _, row in data.iterrows():
    subj = str(row["Subject"])
    rel = str(row["Relation"])
    obj = str(row["Object"])
    
    G.add_node(subj, color="lightblue")  
    G.add_node(obj, color="lightgreen")   
    G.add_edge(subj, obj, label=rel)

net = Network(height="600px", width="100%", directed=True, bgcolor="#222222", font_color="white")

G = nx.relabel_nodes(G, lambda x: str(x))

net.from_nx(G)

for e in G.edges(data=True):
    net.add_edge(e[0], e[1], title=e[2]["label"], label=e[2]["label"])

net.toggle_physics(False)

net.write_html("my_knowledge_graphss.html", open_browser=False)
print("üåê Graph saved as my_knowledge_graphss.html ‚Äî open it in your browser!")

```

```python
import networkx as nx
from pyvis.network import Network

triples = [
    ("Albert Einstein", "developed", "Theory of Relativity"),
    ("Marie Curie", "discovered", "Radium"),
    ("Isaac Newton", "formulated", "Law of Gravitation"),
    ("Charles Babbage", "invented", "Analytical Engine"),
    ("Alan Turing", "pioneered", "Artificial Intelligence"),
    ("Elon Musk", "founded", "SpaceX"),
    ("SpaceX", "launched", "Falcon 9 Rocket"),
    ("Falcon 9 Rocket", "delivered", "Starlink Satellites"),
    ("Google", "created", "TensorFlow"),
    ("TensorFlow", "used for", "Deep Learning")
]

G = nx.DiGraph()

for subj, rel, obj in triples:
    G.add_node(subj, color="lightblue")
    G.add_node(obj, color="lightgreen")
    G.add_edge(subj, obj, label=rel)

net = Network(height="600px", width="100%", directed=True,bgcolor="#222222", font_color="white")

net.from_nx(G)

for e in G.edges(data=True):
    net.add_edge(e[0], e[1], title=e[2]["label"], label=e[2]["label"])

net.toggle_physics(False)

net.write_html("my_knowledge_graph.html", open_browser=False)
print("üåê Graph saved as my_knowledge_graph.html ‚Äî open it in your browser!")
```
